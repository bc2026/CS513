```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./UCI_heart.csv")
df.head()
```

```python
df.dropna(inplace=True)
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt
```

```python
f_cols = ["age", "trestbps", "chol", "thalach"]
for i in range(len(f_cols) - 1):
    for j in range(i + 1, len(f_cols)):
        plt.figure()
        plt.scatter(df[f_cols[i]], df[f_cols[j]], color='blue', alpha=0.5)
        plt.title(f"Scatter Plot of {f_cols[i]} vs {f_cols[j]}")
        plt.xlabel(f_cols[i])
        plt.ylabel(f_cols[j])
        plt.show()

# Box plots for F7, F8, and F9 combinations
import seaborn as sns

f_box_cols = f_cols
for i in range(len(f_box_cols)):
    for j in range(i + 1, len(f_box_cols)):
        plt.figure()
        sns.boxplot(x=df[f_box_cols[j]], y=df[f_box_cols[i]])
        plt.title(f"Box Plot of {f_box_cols[i]} vs {f_box_cols[j]}")
        plt.xlabel(f_box_cols[j])
        plt.ylabel(f_box_cols[i])
        plt.show()
```
