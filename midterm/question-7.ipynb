```python
# Load the data into a pandas dataframe
train_file_path = '/kaggle/input/midterm-uci/Book6.csv'
df_train = pd.read_csv(train_file_path)

test_file_path = '/kaggle/input/midterm-uci/Book7.csv'
df_test = pd.read_csv(test_file_path)
# Remove any rows with missing values 
df_train.dropna(inplace=True) 
df_test.dropna(inplace=True)

df_train.info()
df_test.info()
```

```python
# Select relevant features and target for training data 
features_train = df_train[['age', 'sex', 'trestbps', 'chol']] 
target_train = df_train['Heart_problem'] 
# Select relevant features and target for test data 
features_test = df_test[['age', 'sex', 'trestbps', 'chol']] 
target_test = df_test['Heart_problem']
```

```python
# Standardize the features 
scaler = StandardScaler() 
X_train = scaler.fit_transform(features_train) 
X_test = scaler.transform(features_test)
```

```python
# Initialize and train the KNN classifier with k=3 
knn = KNeighborsClassifier(n_neighbors=3) 
knn.fit(X_train, target_train)
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(target_test, y_pred) 
precision = precision_score(target_test, y_pred) 
recall = recall_score(target_test, y_pred) 
f1 = f1_score(target_test, y_pred) 
# Print Accuracy, Precision, Recall, and F1 Scores 
print(f'Accuracy: {accuracy}%') 
print(f'Precision: {precision}%') 
print(f'Recall: {recall}%') 
print(f'F1 Score: {f1}%')
```

```python
# Print confusion matrix and classification report 
print('Confusion Matrix:') 
print(confusion_matrix(target_test, y_pred)) 
print('Classification Report:') 
print(classification_report(target_test, y_pred))
```
