```python
# Bhagawat Chapagain
# 20014827
# Final Exam
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
# from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler 
```

```python
df = pd.read_csv('./data/UCI_heart.csv')

df.head()
attr = df[['trestbps', 'chol', 'thalach', 'exang']]
scaler = MinMaxScaler()
attr = pd.DataFrame(scaler.fit_transform(attr), columns=attr.columns)
kmeans = KMeans(n_clusters=2, random_state=7)
kmeans.fit(attr)

cluster_labels = kmeans.labels_
clustercenters = kmeans.cluster_centers_
df['Cluster'] = cluster_labels


cluster_0_members = df[df['Cluster'] == 0]
cluster_1_members = df[df['Cluster'] == 1]

print("Members of Cluster 0:")
print(cluster_0_members)


print("\nMembers of Cluster 1:")
print(cluster_1_members)

print("\nCoordinates of Cluster Centers:")
print(clustercenters)
```
